# Setup dependencies

```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import csv
import tensorflow as tf
from tensorflow.keras import layers
```

# Get directories

```python
cwd = os.getcwd()
train_dir = os.path.join(cwd,"train")
test_dir = os.path.join(cwd,"test")
```

# Setup train and test data

```python
names_list = np.array(os.listdir(os.path.join(train_dir, "images")))
n = len(names_list)

names_list
```

```python
xmin = np.empty(n)
xmax = np.empty(n)
ymin = np.empty(n)
ymax = np.empty(n)
```

```python
link = os.path.join(train_dir,"train.csv")
with open(link, 'r+') as file:
    train_labels = csv.DictReader(file)
    i = 0
    
    for row in train_labels:
        xmin[i] = row['xmin']
        xmax[i] = row['xmax']
        ymin[i] = row['ymin']
        ymax[i] = row['ymax']
        i+=1

train_labels1 = [xmin,ymin,xmax,ymax]
train_labels = np.array(train_labels1).transpose()/255
```

```python
train_labels
```

```python
train_img_arr = []
```

```python
for img_name in names_list:
    img_path = os.path.join(train_dir, "images", img_name)
    train_img_arr.append(cv.imread(img_path))
```

```python
train_img_arr = np.array(train_img_arr)
```

```python
(train_img_arr.shape)
```

```python
largest_shape = (0, 0, 0)
for img in train_img_arr:
    if img.shape[0] * img.shape[1] > largest_shape[0] * largest_shape[1]:
        largest_shape = img.shape

print("Largest shape:", largest_shape)
```

```python
test_list = np.array(os.listdir(os.path.join(test_dir, "images")))
m = len(test_list)

test_list
```

```python
test_img_arr = []
for img_name in test_list:
    img_path = os.path.join(test_dir, "images", img_name)
    test_img_arr.append(cv.imread(img_path))
```

```python
for i in range(n):
    x_offset = 0
    y_offset = 0

    blank_image = np.zeros((600,600,3) , dtype='uint8')

    height, width, _ = train_img_arr[i].shape
    
    blank_image[y_offset: y_offset+height, x_offset: x_offset+width] = train_img_arr[i]
    
    x_offset += width

    train_img_arr[i] = blank_image
```

# Building the model

```python
model = tf.keras.Sequential([
    layers.Flatten(input_shape = (600,600,3)),
    layers.Dense(units = 128, activation = 'relu'),
    layers.Dense(units = 4, activation = 'relu')
]
)
```

# Compliling the model

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

# Training the model

```python
# train_img_arr = tf.convert_to_tensor(train_img_arr, dtype='uint8')
print(train_img_arr.shape)
model.fit(train_img_arr, train_labels, epochs = 10)
```
